<h1 align="center", style="color:green">Matplotlib Tutorial: Bar Chart</h1>

### Simple bar chart showing revenues of major US tech companies

```python
company = ['GOOGLE','FB','AMAZONE','MKSF']
revenue= [132,90,144,97]
```

```python
xpos = np.arange(len(company))
xpos
```

```python
plt.bar(xpos,revenue,label='Revenue')

plt.xticks= (xpos,company)
plt.ylabel('Revenue (Bn)')
plt.title("US Stocks")
plt.legend()
```

### Multiple Bars showing revenue and profit of major US tech companies

```python
profit = [40,2,34,12]
```

```python
plt.bar(xpos-0.2,revenue, width=0.4,label='Reveune')
plt.bar(xpos+0.2,profit,width=0.4,label='Profit')

plt.xticks= (xpos,company)
plt.ylabel('Revenue (Bn)')
plt.title("US Stocks")
plt.legend()
```

### Horizontal bar chart using barh function


```python
plt.barh(xpos,revenue,label='Revenue')

plt.xticks= (xpos,company)
plt.ylabel('Revenue (Bn)')
plt.title("US Stocks")
plt.legend()
```

### We have a sample data of blood sugar level of different patients, we will try to plot number of patients by blood range and try to figure out how many patients are normal, pre-diabetic and diabetic

```python
blood_sugar = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
plt.hist(blood_sugar) #by default no.of bins is set to 10
plt.show()
```

<h3 style='color:purple'> bins parameter </h3>

```python
plt.hist(blood_sugar,rwidth=0.5,bins=4)
```

<h3 style='color:purple'>Histogram showing normal, pre-diabetic and diabetic patients distribution</h3>

<ul style="color:brown;font-weight: bold;">
    <li>80-100: Normal</li>
    <li>100-125: Pre-diabetic</li>
    <li>80-100: Diabetic</li>
</ul>

```python
plt.xlabel("Sugar level")
plt.ylabel("No.of Patient")
plt.title("Blood Sugar chart")

plt.hist(blood_sugar,bins=[80,100,125,150],rwidth=0.95,comlor='g')
plt.show()
```

<h3 style='color:purple'>Mutiple data samples in a histogram</h3>

```python
plt.xlabel("Sugar level")
plt.ylabel("No.of Patient")
plt.title("Blood Sugar chart")

blood_sugar_men = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
blood_sugar_women = [67, 98, 89, 120, 133, 150, 84, 69, 89, 79, 120, 112, 100]

plt.hist([blood_sugar_men,blood_sugar_women], bins=[80,100,125,150], rwidth=0.95, color=['green','orange'],label=['men','women'])
plt.legend()
```

<h3 style='color:Green'>histtype=step</h3>

```python
plt.xlabel("Sugar Level")
plt.ylabel("Number Of Patients")
plt.title("Blood Sugar Chart")

plt.hist(blood_sugar,bins=[80,100,125,150],rwidth=0.95,histtype='stepfilled')
```

```python
plt.xlabel("Sugar Level")
plt.ylabel("Number Of Patients")
plt.title("Blood Sugar Chart")

plt.hist(blood_sugar,bins=[80,100,125,150],rwidth=0.95,orientation='horizontal')
```

<h1 align="center", style="color:green">Matplotlib Tutorial: Pie Charts</h1>

<h3 style="color:purple">Draw pie chart to track down home expenses</h3>

```python
exp_vals = [1400,600,300,410,250]
exp_labels = ['Home Rent','Food','Phone/Internet Bill','Car','Other utilities']
plt.pie(exp_vals,labels=exp_labels)
plt.show()
```

<h3 style="color:purple">Draw pie chart as perfect circle</h3>

```python
plt.pie(exp_vals,labels=exp_labels,shadow= True)
plt.axis('equal')
plt.show()
```

<h3 style="color:purple">Show percentages for every pie. Specify radius to increase chart size</h3>

```python
plt.axis("equal")
plt.pie(exp_vals,labels=exp_labels, shadow=True, autopct='%1.f%%',radius=1.5)
plt.show()
```

<h3 style="color:purple">Explode</h3>

```python
plt.axis("equal")
plt.pie(exp_vals,labels=exp_labels, shadow=True, autopct='%1.f%%',radius=1.5,explode=[0,0,0,0.1,0.5])
plt.show()
```


<h3 style="color:purple">counterclock and angle properties</h3>

```python
plt.axis("equal")
plt.pie(exp_vals,labels=exp_labels, shadow=True, autopct='%1.f%%',radius=1.5,explode=[0,0,0,0.1,0.5],counterclock=True,startangle=45)
plt.show()
```

```python
plt.axis("equal")
plt.pie(exp_vals,labels=exp_labels, shadow=True, autopct='%1.f%%',radius=1.5,explode=[0,0,0,0.1,0.5],counterclock=True,startangle=45)
plt.legend()
plt.show()
```

```python
plt.axis("equal")
plt.pie(exp_vals,labels=exp_labels, shadow=True, autopct='%1.f%%',radius=1.5,explode=[0,0,0,0.1,0.5],counterclock=True,startangle=45)
plt.legend(title='Expences:')
plt.show()
```

<h2 style="color:green" align="center">Saving chart to a file using savefig</h2>

```python
plt.axis("equal")
plt.pie(exp_vals,labels=exp_labels, shadow=True, autopct='%1.f%%',radius=1.5,explode=[0,0,0,0.1,0.5])
plt.savefig("piechart.jpg", bbox_inches="tight", pad_inches=2, transparent=True)
```

### You can also save in pdf format and to a different location on your computer by specifying absolute path

```python
plt.pie(exp_vals,labels=exp_labels,radius=2, autopct='%0.1f%%', explode=[0,0.1,0.1,0,0])
plt.savefig("piechart.pdf", bbox_inches="tight", pad_inches=10, transparent=True)
```
